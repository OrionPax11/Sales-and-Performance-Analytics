```python
import pandas as pd
import os
import os
os.makedirs('data', exist_ok=True)

# Load all sheets
df_sales   = pd.read_excel('Internship_Raw_Data.xlsx', sheet_name='Sales_Data')
df_cust    = pd.read_excel('Internship_Raw_Data.xlsx', sheet_name='Customer_Data')
df_prod    = pd.read_excel('Internship_Raw_Data.xlsx', sheet_name='Product_Data')
df_targets = pd.read_excel('Internship_Raw_Data.xlsx', sheet_name='Monthly_Targets')

# Profile
print(df_sales.isnull().sum())
print(df_sales.duplicated().sum())

# Clean
df_sales.dropna(subset=['Order_ID','Order_Date'], inplace=True)
df_sales.drop_duplicates(inplace=True)
df_sales['Region'] = df_sales['Region'].str.strip().str.title()
df_sales['Order_Date'] = pd.to_datetime(df_sales['Order_Date'])

# Feature engineering
df_sales['Month_Name']        = df_sales['Order_Date'].dt.month_name()
df_sales['Quarter']           = 'Q' + df_sales['Order_Date'].dt.quarter.astype(str)
df_sales['Year']              = df_sales['Order_Date'].dt.year
df_sales['Profit_Margin_Pct'] = (df_sales['Profit'] / df_sales['Sales_Amount'] * 100).round(2)

# Merge and export
df_master = df_sales.merge(df_cust, on='Customer_ID', how='left')
df_master = df_master.merge(df_prod, on='Product_Name', how='left')
df_master.to_csv('data/df_master.csv', index=False)
df_targets.to_csv('data/monthly_targets.csv', index=False)
print('Exported. Shape:', df_master.shape)
```
